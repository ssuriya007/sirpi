```python
# @title 1. Data Loading & Cleaning (Initial Exploration)
# Load the dataset using Pandas and check the first five rows.
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Students_Grading_Dataset.csv')
df.head()
```

```python
# Check for missing values and decide how to handle them.
df.isnull().sum()
```

```python
# handling missing values
```

```python
min=df["Attendance (%)"].min()
df["Attendance (%)"]=df["Attendance (%)"].fillna(min)
```

```python
min=df["Assignments_Avg"].min()
df["Assignments_Avg"]=df["Assignments_Avg"].fillna(min)
```

```python
df["Parent_Education_Level"]=df["Parent_Education_Level"].fillna("No Schooling")
```

```python
# Identify data types for each column and convert them if necessary.
df.info()
```

```python
# Remove unnecessary columns (Tip: It could be student personal information like First_Name, Last_Name, Email).
df = df.drop(columns=["First_Name", "Last_Name", "Email"])
print(df)
```

```python
# @title 2. Descriptive Statistics & Understanding Data
# What are the summary statistics (mean, median, std, min, max) of Midterm_Score, Final_Score, and Total_Score?
df[['Midterm_Score', 'Final_Score', 'Total_Score']].describe().T
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Find the age distribution of students in the dataset.
sns.histplot(df["Age"])
plt.show()
```

```python
# Count the number of students in each Department.
# count=df["Department"].value_counts()
# print(count)
df.groupby("Department").size()
```

```python
# What is the average Study Hours per Week for students in different departments?
df["Study_Hours_per_Week"].mean()
```

```python
# @title 3. Academic Performance Analysis
# Which factors correlate most with Total_Score?
correlation_matrix = df.select_dtypes(include=['number']).corr() # Select only numerical columns
total_score_corr = correlation_matrix['Total_Score'].sort_values(ascending=False)
print(total_score_corr)
```

```python
# How does Attendance (%) impact Final Scores?
# (Use a scatter plot to visualize)
plt.scatter(x=df["Attendance (%)"],y= df["Final_Score"])
plt.show()
```

```python
# Does participation in extracurricular activities affect academic performance?
df.groupby("Extracurricular_Activities").size()
```

```python
# Find out if students who study more hours per week score better in exams.
# (Create a visualization to compare Study_Hours_per_Week and Total_Score)
plt.scatter(x=df["Study_Hours_per_Week"],y= df["Total_Score"])
plt.show()
```

```python
# @title 4. Social & Economic Factors Analysis
# Is there a relationship between Parent Education Level and Student Performance?

# Group data by Parent Education Level and calculate average Total Score
parent_education_performance = df.groupby('Parent_Education_Level')['Total_Score'].mean()

# Create a bar plot
sns.barplot(x=parent_education_performance.index, y=parent_education_performance.values)
plt.title('Average Total Score by Parent Education Level')
plt.xlabel('Parent Education Level')
plt.ylabel('Average Total Score')
plt.show()
```

```python
# Do students with Internet Access at Home perform better academically?
internet_access_performance = df.groupby('Internet_Access_at_Home')['Total_Score'].mean()

# Create a bar plot
sns.barplot(x=internet_access_performance.index, y=internet_access_performance.values)
plt.title('Average Total Score by Internet Access at Home')
plt.xlabel('Internet Access at Home')
plt.ylabel('Average Total Score')
plt.show()
```

```python
# Compare the performance of students from different Family Income Levels.
family_income_performance = df.groupby('Family_Income_Level')['Total_Score'].mean()
print(family_income_performance)

# create bar plot
sns.barplot(x=family_income_performance.index, y=family_income_performance.values)
plt.title('Total Score by Family Income Level')
plt.xlabel('Family Income Level')
plt.ylabel('Total Score')
plt.show()
```

```python
# Analyze the effect of Sleep Hours per Night on academic performance.
# (Plot a histogram to understand the distribution)
sleep_hours_performance = df.groupby('Sleep_Hours_per_Night')['Total_Score'].mean()
# print(sleep_hours_performance)
sns.histplot(x=sleep_hours_performance.index,y=sleep_hours_performance.values)
plt.title('Effect of Sleep Hours on Academic Performance')
plt.xlabel('Sleep Hours per Night')
plt.ylabel('Total Score')
plt.show()
```

```python
# @title 5. Stress & Health Impact
# Is there a correlation between Stress Level and Final Scores?
stress_Score=df["Stress_Level (1-10)"].corr(df["Final_Score"])
print(stress_Score)
```

```python
# Find out if students who sleep less than 5 hours per night perform worse than others.
avg_score_less_than_5_hours=df[df["Sleep_Hours_per_Night"]<5]["Total_Score"].mean()
avg_score_atleast_5_hours=df[df["Sleep_Hours_per_Night"]>=5]["Total_Score"].mean()
print("Average Score of student having less than 5 hours sleep: ",avg_score_less_than_5_hours)
print("Average Score of student having atleast 5 hours of sleep ",avg_score_atleast_5_hours)
if(avg_score_less_than_5_hours>avg_score_atleast_5_hours):
  print("Students who sleep less than 5 hours per night perform better than others")
else:
  print("Students who sleep more than 5 hours per night perform better than others")
```

```python
# Identify students who have high stress levels (>7) but high academic performance.
# (What characteristics do they share?)
high_stress_students=df[df["Stress_Level (1-10)"]>7]
avg_totalScore=df["Total_Score"].mean()
# here i have mentioned higher academic performance means more that average total score of all students
print(high_stress_students[high_stress_students["Total_Score"]>avg_totalScore])
```

```python
# @title 6. Visualization & Insights
# Create a boxplot to show the distribution of Total_Score across different departments.
sns.boxplot(x=df["Department"],y=df["Total_Score"])
plt.show()
```

```python
# Use a heatmap to visualize the correlation matrix of numerical columns.
correlation_matrix = df.select_dtypes(include=['number']).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Identify the top 5 students based on Total_Score.
top_5_students = df.nlargest(5, 'Total_Score')
print(top_5_students)
```

```python
# Show the gender distribution in the dataset.
gender_distribution = df['Gender'].value_counts()
gender_distribution.plot(kind='pie', autopct='%.2f')
plt.show()
```
